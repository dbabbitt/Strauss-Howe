```python

tuples_list = [('crisis_year_begin', 'crisis_year_end'), ('turning_year_begin', 'turning_year_end'), ('awakening_year_begin', 'awakening_year_end'),
               ('crisis_begin_year', 'crisis_end_year'), ('birth_year_begin', 'birth_year_end'), ('awakening_begin_year', 'awakening_end_year'),
               ('birth_year', 'death_year'), ('year_of_birth', 'year_of_death'), ('year_reign_begin', 'year_reign_end')]
relevant_pickles_set = set()
for function_name in ['shu.{}'.format(fn) for fn in dir(shu) if fn.endswith('_df')]:
    df = eval(function_name)
    columns_list = df.columns.tolist()
    filters_list = []
    for column_tuple in tuples_list:
        if (column_tuple[0] in columns_list) and (column_tuple[1] in columns_list):
            filters_list.append(f'({function_name}.{column_tuple[0]} <= year) & (year <= {function_name}.{column_tuple[1]})')
            relevant_pickles_set.add(function_name)
    if filters_list:
        print()
        print('mask_series = False\nmask_series = mask_series | (', end='')
        print(')\nmask_series = mask_series | ('.join(filters_list), end='')
        print(')')
        print(f'{function_name}[mask_series].T')
```
